```python
import pandas as pd
```

```python
exam = pd.read_csv('exam.csv')
exam.head()
```

```python
# 1반 (nclass == 1 ) 데이터만 추출 
exam.query('nclass == 1')
```

```python
# 영어점수가 90점 이상인 것만 추출 
result1 = exam.query('english > 90')
result1
```

```python
# 본 파일은 바뀌지 않는다. 
exam
```

```python
# 1반이 아닌 다른 반의 데이터를 추출 
exam.query('nclass != 1')
```

```python
# exam.query('조건 1 & 조건 2 ')
exam.query('nclass == 1 & math >= 50')
```

```python
# exam.query('조건 1 | 조건 2')
exam.query('english < 90 | science < 50 ')
```

```python
# 1반이거나, 3반이거나,5반인 데이터 추출
exam.query('nclass ==1 | nclass == 3 | nclass == 5')
```

```python
# 1반이거나, 3반이거나,5반인 데이터 추출(리스트로 표현)
exam.query('nclass in [ 1 , 3 , 5 ]')
```

```python
# 1반인 데이터 추출하여 데이터 프레임 nclass1로 저장하고 출력
nclass1 = exam.query('nclass == 1')
nclass1
```

```python
# 2반인 데이터 추출하여 데이터 프레임 nclass2로 저장하고 출력
nclass2 = exam.query('nclass == 2')
nclass2
```

```python
nclass2[['math']]
```

```python
# 1반 수학 평균 
nclass1['math'].mean()
```

```python
# 2반 수학 평균 
nclass2['math'].mean()
```

```python
# 전체 반에 대한 평균 
exam['math'].mean()
```

```python
# 2개 변수(sex, country)를 갖는 데이터 프레임 생성
df = pd.DataFrame({'sex' : ['F', 'M', 'F', 'M'],
                   'country' : ['Korea', 'China', 'Japan', 'USA']})
df
```

```python
# sex가 'F'이고 country가 'Korea'인 데이터 추출
# 작은 따옴표와 큰 따옴표에 주의
df.query(' sex == "F" & country == "Korea" ' )
```

```python
# sex가 'F'이고 country가 'Korea'인 데이터 추출
# 작은 따옴표와 큰 따옴표에 주의
df.query(" sex == 'F' & country == 'Korea' " )
```

```python
# 외부 변수 var 값을 query로 사용할 수 있음
n = 'F' 
c = 'Korea'
df.query('sex == @n & country == @c')
```

```python
# 연설문 불러오기 
moon = open('speech_moon.txt', encoding = 'UTF-8').read()
```

```python
moon
```

```python
import re 
moon = re.sub('[^가-힣]', ' ', moon)
moon
```

```python
# hannaum 만들기 
import konlpy
hannanum = konlpy.tag.Hannanum()
```

```python
# 명사 추출 하기 
hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다.")
```

```python
# 연설문에서 명사 추출 하기 
nouns = hannanum.nouns(moon)
nouns
```

```python
# 데이터 프레임으로 변환하기 
import pandas as pd 
df_word = pd.DataFrame({'word' : nouns})
df_word
```

```python
# 단어 빈도표 만들기 
df_word['count'] = df_word['word'].str.len() 
df_word
```

```python
# 두 글자 이상 단어만 남기기 
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
# 단어 빈도 구하기 
df_word = df_word.groupby('word', as_index = False) \
                    .agg(n  = ('word', 'count')) \
                    .sort_values('n', ascending = False)
df_word
```

```python
# 단어 빈도 상위 20개 출력 
top20 = df_word.head(20)
top20
```

```python
# 막대 그래프 만들기 
sns.barplot(data = top20, x = 'n', y = 'word')
```

```python
# 데이터 불러오기 
import pandas as pd 
```

```python
df = pd.read_csv('news_comment_BTS.csv', encoding = 'UTF-8')
```

```python
# 데이터 살펴 보기 
df.info()
```

```python
# 불필요한 문제 제거하기 
df['reply'] = df['reply'].str.replace('[^가-힇]', ' ', regex = True)
df['reply'].head()
```

```python
#명사 추출하기 
# kkma 만들기 
import konlpy 
kkma = konlpy.tag.Kkma()
```

```python
# 명사 추출 - apply() 활용 
nouns = df['reply'].apply(kkma.nouns)
nouns
```

```python
# 한 행에 한 단어가 들어가도록 구성 
nouns = nouns.explode()
nouns
```

```python
# 데이터 프레임 만들기 
df_word = pd.DataFrame({'word' : nouns})

# 글자 수 추가 
df_word['conut'] = df_word['word'].str.len()

# 두글자 이상 단어만 남기기
df_word = df_word.query('conut >= 2 ')
df_word
```

```python
# 빈도표 만들기 
df_word = df_word.groupby('word', as_index = False) \
                .agg(n = ('word', 'count')) \
                .sort_values('n', ascending = False)
df_word
```

```python
# 단어 빈도 상위 20개 추출 
top20 = df_word.head(20)
top20
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams.update({
    'figure.figsize' : [6.5, 6 ]
})
```

```python
# 막대 그래프 만들기 
sns.barplot(data = top20, x = 'n', y = 'word')
```

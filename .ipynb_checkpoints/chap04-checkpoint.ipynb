```python
# Pnadas 패키지를 로드함
# exam.csv를 데이터 프레임 exam으로 읽어들임
import pandas as pd
exam = pd.read_csv('exam.csv')
exam.head()
```

```python
# math 변수만 추출(시리즈)
exam[['math']].head()
```

```python
# nclass, math, english 변수만 데이터 프레임으로 추출
exam[['nclass', 'math', 'english']]
```

```python
# 변수(컬럼) math 제거
#(주의) math 가 제거된 테이블이 반환되어 출력됨 - 원래 데이터프레임은 변경 없음
exam.drop(columns = 'math').head()
```

```python
# exam에서 nclass 가 1인 경우만 추출 
exam.query('nclass == 1')
```

```python
# 2반인 경우만 추출 
exam.query('nclass == 2')
```

```python
# 1반이고 math 점수가 50점 이상인 데이터 추출
exam.query('nclass == 1 & math >= 50')
```

```python
# 1반이 아닌경우 
exam.query('nclass != 1')
```

```python
exam.query('english < 90 | science < 50').head(3)
```

```python
# 1반 , 3반, 5반에 해당하면 추출
exam.query('nclass ==1 | nclass == 3 | nclass == 5')
```

```python
# nclass가 1인 행 추출해 nclass에 할당
nclass1 = exam.query('nclass == 1')
nclass1
```

```python
nclass2 = exam.query('nclass == 2')
nclass2
```

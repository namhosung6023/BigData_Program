```python
# 데이터 프레임 복사본 만들기 
# 변수명을 바꾸기 전에 원본을 보존하기 위해 복사본을 만들고 사본을 작업
mpg_new = mpg.copy()
```

```python
# 고속도로 연비를 나타내는 변수(컬럼) 이름을 hwy에서 highway로 변경
mpg_new = mpg_new.rename(columns = {'hwy' : 'highway'})
mpg_new.head()
```

```python
# 도시 연비를 나타내는 변수(컬럼) 이름을 cty에서 city로 변경
mpg_new = mpg_new.rename(columns = {'cty' : 'city'})
mpg_new.head()
```

```python
# 도시 연비와 고속도로 연비의 평균 값은 통합연비 tatal 파생변수(컬럼) 추가 
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2 
```

```python
mpg.head()
```

```python
mpg['totalNew'] = mpg['total'] + 1 
```

```python
mpg.head()
```

```python
# 자동차 통합 연비 (total) 변수(컬럼)의 요약 통계량 구하기 
mpg['total'].describe()
```

```python
# 통합 연비에 대한 histogram 구하기 
# histogram : 값의 빈도를 막대 길이로 표현 
mpg['total'].plot.hist()
```

```python
# 조건에 따라 서로 다른 값을 부여하기 위한 함수 사용을 위해 numpy 패키지를 로드함 
import numpy as np
```

```python
# np.where()의 사용법 1 
# np.where(조건) => 결과로 조건을 만족하는 index출력
np.where(mpg['total']>=25)
```

```python
# np.where 의 사용법 2 
# np.where(조건, 참일때 반환값, 거짓을때 반환값)
# total 값이 20 이상이면 'pass', 아니면 'fail' 값을 갖는 변수 (컬럼) 'result를 추가함'
mpg['result'] = np.where(mpg['total'] >= 25, 'pass', 'fail')
mpg.head()
```

```python
# 생산년도로 나누기 
# 2001년 이전에 만들어진 자동차의 index를 찾아본다 
np.where(mpg['year'] < 2001)
```

```python
# 생산년도로 나누기 
# trendy라는 컬럼을 만들어서 
# 2001년 이전에 만들어진 자동차는 'old'를 넣고 , 
# 2001년 이후에 생산된 자동차는 'new를 넣는다'
mpg['trendy'] = np.where(mpg['year'] <2001 , 'old', 'new')
```

```python
mpg.head()
```

```python
# result변수는 pass와 fail값을 가지며 각 분류 값을 갖는 데이터 빈도수 

count_test = mpg['result'].value_counts()
count_test
```

```python
# pass와 fail값을 가지고 있는 coount_test변수의 막대그래프 만들기 
count_test.plot.bar()
```

```python
count_test.plot.pie()
```

```python
# 글자 위치 조정하기 
count_test.plot.bar(rot=0)
```

```python
# 통합 연비 값에 따라 A, B, C 등급으로 분류한 값을 갖는 'grade' 변수 추가
mpg['grade'] = np.where(mpg['total'] >= 30, 'A',
                        np.where(mpg['total'] >= 20, 'B', 'C'))
mpg.head()
```

```python
# 통합 연비 값에 따라 A B C 등급으로 분류되니 데이터 개수 테이블 생성
count_grade = mpg['grade'].value_counts()
count_grade
```

```python
# count_grade의 등급 빈도 막대그래프 그리기 
count_grade.plot.bar(rot=0)
```

```python
midwest = pd.read_csv('midwest.csv')
```

```python
midwest.head()
```

```python
midwest.shape
```

```python
midwest.info()
```

```python
# poptotal(전체 인구) 변수를 total로 변경
midwest = midwest.rename(columns={'poptotal' : 'total'})
midwest.head()
```

```python
# popasian (아시아 계 인구수) 변수를 asian으로 수정 
midwest = midwest.rename(columns={'popasian': 'asian'})
midwest.head()
```

```python
# 결측지, 데이터 타입, 데이터 개수 등 파악 
midwest.info()
```

```python
# 전체 인구(total) 대비 안시아 인구 (asian) 백분률 구하기 
midwest['rate'] = (midwest['asian'] / midwest['total']) * 100
midwest.head()
```

```python
# 데이터 프레임 midwest에서 ['county', 'state', 'total', 'asian','rate'] 컬럼(변수)
midwest[['county', 'state', 'total', 'asian', 'rate']].head(11)
```

```python
# 전체 인구(total) 대비 안시아 인구(asian) 백분율 빈도수 확인
midwest['rate'].plot.hist()
```

```python
# 데이터 프레임 midwest의 전체 인구(total) 대비 안시아 인구(asian) 백분율 요약 통계량
midwest['rate'].describe()
```

```python
# 데이터 프레임 midwest의 전체 인구(total) 대비 안시아 인구(asian) 백분율 평균 구하기
midwest['rate'].mean()
```

```python
# 전체 인구(total) 대비 안시아 인구(asian) 백분율
# 평균 값 보다 크면 large 작으면 small 값을 갖는 파생 변수 group 생성
midwest['group'] = np.where(midwest['rate'] > 0.4872, 'large', 'small') # 평균 값
midwest['group'].head()
```

```python
# 변수 group의 값에 따른 빈도수 출력
midwest['group'].value_counts()
```

```python
# 변수 group의 값에 따른 빈도수 표를 막대 그래프로 작성
count_group = midwest['group'].value_counts().plot.bar(rot= 0)
```
